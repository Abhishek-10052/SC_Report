```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Input, Dense
from keras.models import Model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from keras.models import Model
from tensorflow import keras
from keras.layers import LeakyReLU, Input, Dense
```

```python
my_data = pd.read_csv("/content/drive/MyDrive/training_pixel_data_classsfication1.csv")
my_data.head()
```

```python
my_data.drop(['OBJ_Id'], axis = 1, inplace = True)
```

```python
my_data.head()
```

```python
my_data.drop(['OBJ_Cl'], axis = 1, inplace = True)
```

```python
my_data.head()
```

```python
# Normalize the reflectance data to a range of 0 to 1
normalized_data = (my_data + 1) / 3.0
```

```python
# Autoencoder architecture
input_dim = normalized_data.shape[1]  # Number of bands
encoding_dim = 20# Dimensionality of the compressed representation
```

```python
input_data = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation=LeakyReLU(alpha=0.3))(input_data)
decoded = Dense(input_dim, activation='relu')(encoded)
```

```python
autoencoder = Model(input_data, decoded)
```

```python
# Compile the model
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the autoencoder
autoencoder.fit(normalized_data, normalized_data, epochs=53, batch_size=26)
```

```python
# Get the encoder part of the autoencoder
encoder = Model(input_data, encoded)
```

```python
# Use encoder to get the encoded representation of the data
encoded_data = encoder.predict(normalized_data)
```

```python
labels = np.random.randint(0, 2, size=1190)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(encoded_data, labels, test_size=0.2, random_state=42)
```

```python
# Train a classifier (e.g., Random Forest) using the encoded representation
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)


```

```python
# Get feature importances from the classifier
feature_importances = clf.feature_importances_
```

```python
# Rank features based on their importances
ranked_features = np.argsort(feature_importances)[::-1]
```

```python
# Print ranked features
for rank, feature_index in enumerate(ranked_features):
    print(f"Rank {rank + 1}: Feature {feature_index}, Importance: {feature_importances[feature_index]}")
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Evaluate the classifier on training data
y_train_pred = clf.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
```

```python
# Evaluate the classifier on testing data
y_test_pred = clf.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report
```

```python
# Print classification report for detailed performance metrics
print("Classification Report on Testing Data:")
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn import metrics
```

```python
# Calculate confusion matrix
conf_matrix = metrics.confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Extract True Positives, True Negatives, False Positives, and False Negatives
tn, fp, fn, tp = conf_matrix.ravel()
print(f"True Positives (TP): {tp}")
print(f"True Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
```
